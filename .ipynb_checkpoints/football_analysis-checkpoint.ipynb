```python
import pandas as pd
import requests
import matplotlib.pyplot as mp
```

```python
df=pd.read_csv('champions.csv')
df
```

```python
df.describe()
```

### selecting only team and titles 


```python
winners=df.loc[:,['Team','Titles']]
winners
```

### plotting most no of worldcup winners 

```python
winners.plot(kind='bar',x='Team',y='Titles')
```

```python
games=df.loc[:,['Team','GF','GA']]
games
```

### Goal For and Goal Against in a bar graph representation by a team


```python
games.plot(kind='bar',y=['GF','GA'],x='Team',color=['yellow','green'],figsize=(6,10),)
```

```python
my_team=df.Team
my_pts=df.Pts
```

### showing total points gained by a team as percentage in the form of pie chart


```python
mp.pie(my_pts,labels=my_team,autopct='%1.0f%%')
mp.title("pie chart to show total points gained")
mp.show()
```
